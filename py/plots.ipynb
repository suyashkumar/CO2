```python
import json
import seaborn
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.DataFrame()
with open('data.ndjson', 'r') as f:
    for line in f.readlines():
        new_data = pd.DataFrame([json.loads(line)])
        data = pd.concat([data, new_data])
    
```

```python
data["time"] = pd.to_datetime(data["time"])
data = data.sort_values(by="time")
```

```python
data
```

```python
LAST_N = 800
```

```python
sampled = data.sample(frac=0.5)
data_last_n = data.tail(LAST_N)
```

```python
plt.figure(1, figsize=(10, 5))
seaborn.lineplot(data_last_n, x="time", y="co2_ppm")
plt.title(f"CO2 PPM (Last {LAST_N} measurements)")
```
